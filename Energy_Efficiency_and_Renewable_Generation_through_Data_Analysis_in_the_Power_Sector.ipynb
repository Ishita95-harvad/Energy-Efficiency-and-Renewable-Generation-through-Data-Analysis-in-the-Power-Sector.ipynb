```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming 'df' is your DataFrame
# ... (your existing code to create and process 'df') ...

# 'solar_gen' seems to be the intended target variable
X = df.drop(columns=['solar_gen'])
y = df['solar_gen']

# Remove duplicate column names (keeping the first occurrence)
X = X.loc[:,~X.columns.duplicated()]

# --- Perform train-test split before feature engineering ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# --- Now proceed with feature engineering on the training and testing sets separately ---

# Create a ColumnTransformer to one-hot encode 'region' and scale numerical features
categorical_features = ['region']  # List of categorical columns
numerical_features = X.select_dtypes(exclude=['object', 'datetime64[ns]']).columns.tolist() # Select numerical features, excluding datetime

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),  # Scale numerical features
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features),  # One-hot encode categorical features
    ])

# Create a pipeline to combine preprocessing and model training
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Train the pipeline
pipeline.fit(X_train, y_train) # Use encoded training data

# Predictions
y_pred = pipeline.predict(X_test) # Use encoded testing data

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae:.2f}')
print(f'R2 Score: {r2:.2f}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense

# Instead of loading from a CSV, use your existing DataFrame 'df'
# and extract the 'Total_Generation' column
data = df['Total_Generation'].values.reshape(-1, 1)  # Reshape to (samples, features)

# Scale data to the range [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data)

# Split data into train and test sets (80% for training, 20% for testing)
train_size = int(len(data) * 0.8)
train_data, test_data = data[0:train_size, :], data[train_size:len(data), :]

# Create a function to prepare sequences for LSTM/GRU
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        X.append(dataset[i:(i+look_back), 0])  # Features
        Y.append(dataset[i + look_back, 0])  # Target
    return np.array(X), np.array(Y)

# Choose a look_back period (e.g., 7 days)
look_back = 7
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

# Reshape input to be [samples, time steps, features] for LSTM
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model_lstm.add(LSTM(units=50))
model_lstm.add(Dense(1))  # Output layer

# Compile the model
model_lstm.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model_lstm.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

# Make predictions
predictions_lstm = model_lstm.predict(X_test)

# Invert scaling to get actual values
predictions_lstm = scaler.inverse_transform(predictions_lstm)
y_test = scaler.inverse_transform([y_test])


# Plot predictions against actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test[0], label='Actual')
plt.plot(predictions_lstm, label='LSTM Predictions', linestyle='--')
plt.title('LSTM Energy Generation Forecast vs Actual')
plt.xlabel('Time Step')
plt.ylabel('Energy Generation (MW)')
plt.legend()
plt.show()
```

**9. Hyperparameter Tuning**

Use GridSearchCV or RandomizedSearchCV to optimize model hyperparameters.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming 'df' is your DataFrame with features and target variable

# 'solar_gen' seems to be the intended target variable
# based on previous cells and the DataFrame structure
X = df.drop(columns=['solar_gen'])  # Features (excluding the target variable)
y = df['solar_gen']  # Target variable

# Remove duplicate column names (keeping the first occurrence)
X = X.loc[:,~X.columns.duplicated()]

# --- Perform train-test split before feature engineering ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# --- Now proceed with feature engineering on the training and testing sets separately ---

# Create a ColumnTransformer to one-hot encode 'region' and scale numerical features
categorical_features = ['region']  # List of categorical columns
numerical_features = X.select_dtypes(exclude=['object', 'datetime64[ns]']).columns.tolist() # Select numerical features, excluding datetime

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),  # Scale numerical features
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features),  # One-hot encode categorical features
    ])

# Create a pipeline to combine preprocessing and model training
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=42)) # Removed n_estimators from here
])


# Hyperparameter tuning for Random Forest, now using the pipeline
param_grid = {
    'regressor__n_estimators': [50, 100, 200], # Adjusted parameter names for pipeline
    'regressor__max_depth': [10, 20, 30],
    'regressor__min_samples_split': [2, 5, 10],
    'regressor__min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=3, n_jobs=-1) # Using pipeline in GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters
print(f'Best Parameters: {grid_search.best_params_}')
```
